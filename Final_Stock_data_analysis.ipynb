#Task 2: Strategy development-
Here's I developed a simple moving average crossover strategy for one of the stocks from our list, backtest it, and compare the returns against a simple buy-and-hold approach.

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the stock and data  for the last year
stock = 'RELIANCE.NS'
data = yf.download(stock, start='2023-09-01', end= '2024-09-01')
```

```python
data.head()
```

```python
# calculate the simple moving averages
data['SMA_50']= data['Adj Close'].rolling(window=50).mean()
data['SMA_200']= data['Adj Close'].rolling(window=200).mean()
```

```python
data['SMA_50'].head()
```

```python
data['SMA_200'].head()
```

```python
# Generate buy and sell signals
data['Signal'] = 0
data['Signal'] = np.where(data['SMA_50'] > data['SMA_200'], 1, 0)  # 1 for buy, 0 for sell
data['Position'] = data['Signal'].diff()  # 1 for buy, -1 for sell
```

```python
# Backtesting the strategy
# Calculate returns based on signals
data['Market_Returns'] = data['Adj Close'].pct_change()
data['Strategy_Returns'] = data['Market_Returns'] * data['Signal'].shift(1)
```

```python
# Calculate cumulative returns
data['Cumulative_Market_Returns'] = (1 + data['Market_Returns']).cumprod()
data['Cumulative_Strategy_Returns'] = (1 + data['Strategy_Returns']).cumprod()
```

```python
# Plotting the results
plt.figure(figsize=(14, 7))
```

```python
# Plot the stock prices and moving averages
plt.subplot(2, 1, 1)
plt.plot(data['Adj Close'], label='Adjusted Close Price', alpha=0.5)
plt.plot(data['SMA_50'], label='50-Day SMA', alpha=0.75)
plt.plot(data['SMA_200'], label='200-Day SMA', alpha=0.75)
plt.title(f'{stock} Price and Moving Averages')
plt.legend()
plt.grid()
```

```python

# Plot the cumulative returns
plt.subplot(2, 1, 2)
plt.plot(data['Cumulative_Market_Returns'], label='Buy and Hold Strategy', alpha=0.75)
plt.plot(data['Cumulative_Strategy_Returns'], label='Moving Average Crossover Strategy', alpha=0.75)
plt.title('Cumulative Returns')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Print final returns
final_market_return = data['Cumulative_Market_Returns'].iloc[-1] - 1
final_strategy_return = data['Cumulative_Strategy_Returns'].iloc[-1] - 1
```

```python

print(f"Final Market Return: {final_market_return:.2%}")
print(f"Final Strategy Return: {final_strategy_return:.2%}")
```

# Q2. Task 3: Risk management-
To manage risk in our SMA crossover strategy, we’ll calculate the maximum drawdown for each stock and implement a stop-loss mechanism. The steps below explain the process.

1. Maximum Drawdown Calculation-
The maximum drawdown is the largest drop from a peak to a trough for a stock over a specific period. It helps assess the risk involved in holding the stock.

We will calculate the maximum drawdown for each stock over the 6-month period.

```python
import numpy as np

# Function to calculate maximum drawdown
def calculate_max_drawdown(data):
    # Calculate the cumulative maximum of the stock prices
    data['Cumulative Max'] = data['Close'].cummax()
    # Calculate drawdown (percentage drop from peak)
    data['Drawdown'] = (data['Close'] - data['Cumulative Max']) / data['Cumulative Max']
    # Calculate maximum drawdown
    max_drawdown = data['Drawdown'].min()
    return max_drawdown
```

```python
# Example: Calculate maximum drawdown for 5 stocks from Nifty 50
stocks = ['RELIANCE.NS', 'TCS.NS', 'INFY.NS', 'KOTAKBANK.NS', 'ICICIBANK.NS']
max_drawdowns = {}

for stock in stocks:
    data = yf.download(stock, period='6mo')
    max_drawdown = calculate_max_drawdown(data)
    max_drawdowns[stock] = max_drawdown
```

```python
# Display the maximum drawdown for each stock
print("Maximum Drawdown for each stock:")
for stock, drawdown in max_drawdowns.items():
    print(f"{stock}: {drawdown:.2%}")
```

Analysis:
The stock with the highest drawdown (most significant decline from its peak) is KOTAKBANK.NS at -16.07%.
The stock with the lowest drawdown is ICICIBANK.NS at -7.65%.

Step 2: Implementing the Stop-Loss Mechanism

Steps:
Calculate original strategy returns using the 50-day and 200-day SMA crossover.
Implement a 5% stop-loss mechanism to exit trades if the price falls by 5% after entering.
Compare the performance of the two strategies.

```python
# Fetch historical stock data for Reliance (6 months)
data = yf.download('RELIANCE.NS', period='1y')
# Calculate 50-day and 200-day SMAs
data['SMA50'] = data['Close'].rolling(window=50).mean()
data['SMA200'] = data['Close'].rolling(window=200).mean()
```

```python

# Drop rows where SMA values are NaN (before the SMAs can be calculated)
data = data.dropna(subset=['SMA50', 'SMA200'])

# Generate Buy/Sell signals using .loc to avoid SettingWithCopyWarning
data.loc[50:, 'Signal'] = np.where(data['SMA50'][50:] > data['SMA200'][50:], 1, 0)

# Calculate the difference in signals to determine position changes (Buy/Sell)
data['Position'] = data['Signal'].diff()
```

```python
# Function to backtest with or without stop-loss
def backtest_with_stop_loss(data, stop_loss_pct=0.05):
    """
    Backtest the SMA crossover strategy with a stop-loss mechanism.
    :param stop_loss_pct: The stop-loss percentage (e.g., 0.05 for 5%)
    :return: Total return from the strategy
    """
    # Generate Buy/Sell signals
    data['Signal'] = 0
    data['Signal'][50:] = np.where(data['SMA50'][50:] > data['SMA200'][50:], 1, 0)
    data['Position'] = data['Signal'].diff()

    # Variables to track performance
    entry_price = None
    stop_loss_price = None
    strategy_returns = []

     # Loop through each row of data to simulate the strategy
    for i in range(1, len(data)):
        # If there's a buy signal and no open position, enter the trade
        if data['Position'].iloc[i] == 1 and entry_price is None:
            entry_price = data['Close'].iloc[i]
            stop_loss_price = entry_price * (1 - stop_loss_pct)  # Set stop-loss price

        # If a position is open, check if stop-loss or sell signal is triggered
        elif entry_price is not None:
            # Check if stop-loss is triggered
            if data['Close'].iloc[i] < stop_loss_price:
                strategy_returns.append((data['Close'].iloc[i] - entry_price) / entry_price)
                entry_price = None
                stop_loss_price = None
            # Check for sell signal (crossover or exit condition)
            elif data['Position'].iloc[i] == -1:
                strategy_returns.append((data['Close'].iloc[i] - entry_price) / entry_price)
                entry_price = None
                stop_loss_price = None

    # Calculate total strategy return
    total_return = np.sum(strategy_returns)
    return total_return
```

```python
# Calculate original strategy returns (without stop-loss)
original_strategy_return = backtest_with_stop_loss(data, stop_loss_pct=0)  # No stop-loss

# Calculate strategy returns with a 5% stop-loss
stop_loss_strategy_return = backtest_with_stop_loss(data, stop_loss_pct=0.05)

# Compare the returns
print(f"Original Strategy Return (No Stop-Loss): {original_strategy_return:.2%}")
print(f"Strategy Return with 5% Stop-Loss: {stop_loss_strategy_return:.2%}")
```
