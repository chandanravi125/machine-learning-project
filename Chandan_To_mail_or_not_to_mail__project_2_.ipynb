# 1. Handling Imbalanced Data
Imbalanced datasets occur when one class significantly outnumbers the others. Techniques like SMOTE or class weighting help in these scenarios.
* A.  SMOTE generates synthetic samples for the minority class to balance the dataset.
Use SMOTE after splitting data into training and test sets.

* B. Class Weighting
Many machine learning models, like logistic regression or random forests, allow for class weights.
This technique assigns higher weights to the minority class during training.
#


# 2. Feature Engineering
Feature engineering helps extract more information from the data to improve model performance.

* A. Identify and Drop Low-Variance Features
Low-variance features carry little information and may be noise. Use variance thresholds to identify such features.

* B. Add Interaction Terms
Interaction terms represent the product of two or more features, capturing relationships between them.

* C. Create Domain-Specific Features
Use domain knowledge to craft features relevant to the problem.
For example, in healthcare, BMI (weight/height²) is a meaningful feature derived from height and weight.
* D. Normalize or Standardize Features
If features have different scales, normalize or standardize them.


```python
# Separate features and target
X = train_df.drop('V86', axis=1)
y = train_df['V86']
#scaler = StandardScaler()
#X = scaler.fit_transform(X)
# Split data into training and validation sets
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

# 1. Handle Imbalance using SMOTE
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# 2. Feature Engineering
# Drop low-variance features
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(threshold=0.01)
X_train_resampled = selector.fit_transform(X_train_resampled)
X_val = selector.transform(X_val)

# Add interaction terms
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_train_resampled = poly.fit_transform(X_train_resampled)
X_val = poly.transform(X_val)

# Normalize data
scaler = StandardScaler()
X_train_resampled = scaler.fit_transform(X_train_resampled)
X_val = scaler.transform(X_val)

# 3. Train the model
model = LogisticRegression(class_weight='balanced', random_state=42)
model.fit(X_train_resampled, y_train_resampled)

# 4. Evaluate
y_pred = model.predict(X_val)
from sklearn.metrics import fbeta_score
# Calculate F-beta score with beta = 2 on validation set
fbeta_val = fbeta_score(y_val, y_pred, beta=2)

print("F-beta score on validation set (beta=2):", fbeta_val)
```

Hyperparameter Tuning
Common Parameters:
* Learning rate (learning_rate): Controls step size during training.
* Number of estimators (n_estimators): Total boosting rounds.
* Max depth (max_depth): Depth of each tree.
* Minimum data in leaf (min_child_samples for LightGBM): Prevents overfitting.
* Scale pos weight (scale_pos_weight): Balances classes for imbalanced datasets.

```python
# Grid Search for LightGBM
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5, 7],
    'scale_pos_weight': [1, 5, 10]  # Adjust for imbalance
}

grid = GridSearchCV(LGBMClassifier(random_state=42), params, cv=3, scoring='f1')
grid.fit(X_train, y_train)

print(f"Best parameters: {grid.best_params_}")
print(f"Best score: {grid.best_score_}")
```
